```python
patient_records = {
1: {"Name": "Ali", "Age": 45, "Diagnosis": "Flu"},
2: {"Name": "Ayse", "Age": 30, "Diagnosis": "High Blood Pressure"},
3: {"Name": "Mehmet", "Age": 60, "Diagnosis": "Diabetes"},
4: {"Name": "Ahmet", "Age": 50, "Diagnosis": "Heart Disease"}
}

print(""" 
    Hospital Automation System
    1. Add New Patient
    2. View Patient Records
    3. Search for a Specific Patient
    4. Delete Patient Record
    5. Exit""")

while True:
    choice = input("Please select an action (1-5): ")
    
    if choice == "1":
        patient_no = int(input("Enter patient number: "))
        Name = input("Enter patient name: ")
        Age = int(input("Enter patient age: "))
        Diagnosis = input("Enter diagnosis: ")
        patient_records[patient_no] = {"Name": Name, "Age": Age, "Diagnosis": Diagnosis}
        print(f"New patient added: {patient_no} - {Name}")
        
    elif choice == "2":
        print("Current Patients")
        print(f"{'Number'.ljust(10)} | {'Name'.ljust(20)} | {'Age'.ljust(5)} | {'Diagnosis'.ljust(20)}")
        print("-" * 60)
        for patient_no, patient_info in patient_records.items():
            print(f"{str(patient_no).ljust(10)} | {patient_info['Name'].ljust(20)} | {str(patient_info['Age']).ljust(5)} | {patient_info['Diagnosis'].ljust(20)}")
        
    elif choice == "3":
        patient_name = input("Enter the name of the patient you want to view: ")
        found = False
        for patient_no, patient_info in patient_records.items():
            if patient_info["Name"] == patient_name:
                print(f"Number: {patient_no} | Name: {patient_info['Name']} | Age: {patient_info['Age']} | Diagnosis: {patient_info['Diagnosis']}")
                found = True
                break
        if not found:
            print(f"No patient found with the name {patient_name}")
             
    elif choice == "4":
        patient_no = int(input("Enter the number of the patient you want to delete: "))
        if patient_no in patient_records:
            patient_name = patient_records[patient_no]["Name"]
            del patient_records[patient_no]
            print(f"Deleted patient number: {patient_no}, name: {patient_name}")
        else:
            print(f"No patient found with the number {patient_no}")
            
    elif choice == "5":
        print("Exiting Patient Record System")
        break
    else:
        print("Invalid choice, please try again.")
```
